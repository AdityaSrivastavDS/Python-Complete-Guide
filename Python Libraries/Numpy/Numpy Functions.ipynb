```python
import numpy as np
```

```python
#where() is used to search elements
#returns index 
var = np.array([1,2,3,4,5,2,6,5])

search = np.where(var == 5) #pass the variable and elements which you wanted to find

print(search)
```

```python
var = np.array([1,2,3,4,5,2,6,5])

search = np.where(var%2 == 0) #we can pass with conditions as well

print(search)
```

```python
#searchsorted() is used to search in sorted array
arr = np.array([1,2,3,4,5,6])

result = np.searchsorted(arr, 5)

print("The given element is present at index:- ",result)
```

```python
#we can also change the direction of searching using side with searchsorted
arr = np.array([1,2,3,4,5,6])

result = np.searchsorted(arr, 5, side="right")  #the search will start from right side

print("The given element is present at index:- ",result)
```

```python
#we can search multiple element as well
arr = np.array([1,2,3,4,5,6])

result = np.searchsorted(arr, [3,4,5])

print("The given element is present at index:- ",result)
```

```python
#sort() is used for sorting array
arr1 = np.array([3,7,1,6,12,9,2])

print("Original array:- ",arr1)
print()

print("Sorted array is:- ", np.sort(arr1))
```

```python
var1 = np.array(['d','a','c','b'])

print(np.sort(var1))
```

```python
#sorting 2-d array
arr2 = np.array([[3,7,1],[6,12,9]])

print("Sorted 2-d array :-")
print(np.sort(arr2))
```

```python
filt_arr = np.array(['d','a','c','b'])

filt = [True,False,True,False]

new_arr = filt_arr[filt]

print(new_arr)
```

***shuffle:***

```python
#random.shuffle() arrange the sequence in random order
var = np.array([1,2,3,4,5])

np.random.shuffle(var)

print(var)
```

***unique:***

```python
#unique() returns unique data from the sequence
var1 = np.array([1,2,3,4,2,5,6,7])

res = np.unique(var1)

print(res)
```

```python
var1 = np.array([1,2,3,4,2,5,6,7])

res = np.unique(var1, return_index = True) #return_index = True returns index of all unique numbers

print(res)
```

```python
var1 = np.array([1,2,3,4,2,5,6,7])

res = np.unique(var1, return_index = True, return_counts = True) #return_counts = True returns counting of all unique numbers

print(res)
```

***resize:***

```python
#resize() used to change the size of existing array
var2 = np.array([1,2,3,4,5,6])

y = np.resize(var2,(2,3))

print(y)
```

***flatten***

```python
#converts multidimensional array into 1-d array
var2 = np.array([[1,2,3],[4,5,6]])

print(var2.flatten())
```

```python
var2 = np.array([[1,2,3],[4,5,6]])

print(var2.flatten(order="F"))
```
