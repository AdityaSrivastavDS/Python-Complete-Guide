```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
var = [1,2,3,4,5,6,7]
var_1 = [2,3,4,1,5,6,7]

#creating DataFrame as seaborn works on onlyb datasets
x = pd.DataFrame({"var":var, "var_1":var_1})


sns.lineplot(x = "var", y = "var_1", data = x)


plt.show()
```

```python
#reading data from datset
data = pd.read_csv("C:\\Users\\PC\\Downloads\\JamesBondMovies.csv")
```

```python
#observing data using head()
#by default its range is 5
data.head()
```

```python
#Plotting data

#we will decide the size of figure, its not necessary but defined as standard practice
#for this we will use figure() and its figsize parameter
plt.figure(figsize=(16,6))


#plotting line plot
'''lineplot() takes three parameter
        1.x = name of column which we wanted to make x axis
        2.y = column name which we wanted to make y axis
        3.data parameter in which we will pass the variable in which our dataset is stored'''

sns.lineplot(x = "Budget", y = "Box Office", data = data)
```

```python
#changing the color of line using color parameter
sns.lineplot(x = "Budget", y = "Box Office", data = data, color = "magenta")
```

```python
#changing width of line using linewidth parameter
sns.lineplot(x = "Budget", y = "Box Office", data = data, linewidth = 5)

plt.show()
```

```python
#adjusting background color
#for formatting seaborn pots we use set() function
'''1.rc parameter is used for formatting 
   2.we pass all styles in form of dictionary within rc
   3.for background color we use axes.facecolor:color_name'''

sns.set(rc = {"axes.facecolor":"black"})
sns.lineplot(x = "Budget", y = "Box Office", data = data, color = "red")

plt.show()
```

```python
'''whenever we changes background color of plot it automaticaly set to grid form by default
   
   To overcome this problem we will set axes.grid to False'''

sns.set(rc = {"axes.facecolor":"black","axes.grid":False})
sns.lineplot(x = "Budget", y = "Box Office", data = data, color = "red")

plt.show()
```

```python
'''changing label size:
    1.xtick.labelsize for x-axis label
    2.ytick.labelsize for y-axis label''' 

sns.set(rc = {"xtick.labelsize":15, "ytick.labelsize":16})
sns.lineplot(x = "Budget", y = "Box Office", data = data, color = "red")

plt.show()
```

```python
#giving title to plot using title()
plt.figure(figsize=(16,6))

sns.lineplot(x = "Budget", y = "Box Office", data = data)

plt.title("James Bond Movies History")

plt.show()
```

```python
#labelling to the line of legend using label parameter
plt.figure(figsize=(16,6))

sns.lineplot(x = "Budget", y = "Box Office", data = data, label = "Collection and Investment")

plt.show()
```

```python
# Using markers to identify groups
sns.set_style('darkgrid')
sns.lineplot(x = "Budget",y = "Box Office", err_style = "bars", markers = True, data = data)

plt.show()
```

```python
sns.lineplot(x = "Budget", y = "Box Office", dashes = False, palette = "CMRmap", data = data)

sns.set(style = "dark")


plt.show()
```
