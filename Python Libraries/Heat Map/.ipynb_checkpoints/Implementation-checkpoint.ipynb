```python
#Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading Data
data = [
    [0.2, 0.4, 0.6, 0.8],
    [0.3, 0.5, 0.7, 0.9],
    [0.1, 0.2, 0.3, 0.4],
    [0.7, 0.6, 0.5, 0.4]
]
```

```python
#plotting heatmap
sns.heatmap(data = data, annot = True)
plt.show()
```

```python
#changing color of heatmap using cmap parameter
#plotting heatmap
sns.heatmap(data = data, annot = True, cmap = "YlGnBu")
plt.show()
```

```python
#removing color bar from heatmap using cbar parameter
sns.heatmap(data = data, annot = True, cmap = "YlGnBu", cbar = False)
plt.show()
```
