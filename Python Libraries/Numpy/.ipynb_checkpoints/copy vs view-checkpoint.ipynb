**Copy**

•A copy of an array is a new array with its own data that is completely independent of the original array.

•Modifying the copy will not affect the original array, and vice versa.

•The ***copy()*** method is used to create a deep copy of the array.

```python
import numpy as np
```

```python
var = np.array([1,2,3,4])

co = var.copy()

print("First array:- ",var)
print("Copied array:- ",co)
```

**View**

•A view of an array is a new array object that refers to the same underlying data as the original array.

•Modifying the view will affect the original array, and vice versa.

•Views are created using slicing, which produces a new array that is a "view" into the original array's data without copying it.

```python
#there is two way to create view
#first is that we can use view() function
arr = np.array([9,6,67,78,45])

view_arr = arr.view()

print("Original array:- ",arr)
print("View array:- ",view_arr)
```

```python
#second method is that we can use slicing to create view
arr1 = np.array([6,3,1,7,9,6,4])

view_arr1 = arr1[0:4]

print("Original array:- ",arr1)
print("View array:- ",view_arr1)
```
