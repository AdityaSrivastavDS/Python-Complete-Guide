```python
import pandas as pd
```

**Merge**

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2 = pd.DataFrame({"A":[1,2,3,4],"B1":[21,22,23,24]})
```

```python
var1
```

```python
var2
```

```python
#merge() TO MERGE TWO DATAFRAMES
#both dataframes must have at least one identical ID
#It takes three parameter:- first_df:second_df:on="common_id"
pd.merge(var1,var2,on = "A")
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2 = pd.DataFrame({"A":[1,2,3,5],"C":[21,22,23,24]})
```

```python
pd.merge(var1,var2)
```

```python
#Getting missing values of lef side
pd.merge(var1,var2, how="left")
```

```python
#Getting missing values of right side
pd.merge(var1,var2, how="right")
```

```python
#Getting all missing values
pd.merge(var1,var2, how="outer")
```

```python
#we can see that which data get merged through indicator=True
pd.merge(var1,var2, how="outer",indicator=True)
```

```python
var1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

var2 = pd.DataFrame({"A":[1,2,3,5],"B":[21,22,23,24]})
```

```python
pd.merge(var1,var2)
```

```python
#In above merge values are not visible
#To make it visible we will pass parameter  left_index and right_index as True

pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
#changing name of columns using suffixes
pd.merge(var1,var2,left_index=True,right_index=True, suffixes=("Pandas","Python"))
```

**Concat**

```python
sr1 = pd.Series([1,2,3,4])
sr2 = pd.Series([11,21,31,41])
```

```python
sr1
```

```python
sr2
```

```python
pd.concat([sr1,sr2])
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

df2 = pd.DataFrame({"A":[1,2,3,4],"B":[21,22,23,24]})
```

```python
pd.concat([df1,df2])
```

```python
#concatenating along axis
pd.concat([df1,df2],axis=0)
```

```python
pd.concat([df1,df2],axis=1)
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

df2 = pd.DataFrame({"A":[1,2],"B":[21,22]})
```

```python
pd.concat([df1,df2],axis=1)
```

```python
#dealing with missing values
pd.concat([df1,df2],axis=1, join="inner")
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4],"B":[11,12,13,14]})

df2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,32,41]})
```

```python
#merging data along key
pd.concat([df1,df2],axis=1,keys=["d1","d2"])
```

```python
pd.concat([df1,df2],axis=0,keys=["d1","d2"])
```

```python
df1 = pd.DataFrame({"A":[1,2,3,4]})

df2 = pd.DataFrame({"A":[1,2,3,4],"C":[21,22,32,41]})
```

```python
#merging dataframes of two different dimensions
pd.concat([df1,df2])
```
