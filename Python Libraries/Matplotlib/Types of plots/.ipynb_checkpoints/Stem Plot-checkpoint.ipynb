**Stem Plot**

•A stem plot is a type of plot used to visualize discrete data. It is often used to display values along a number line. 

•In a stem plot, each data point is represented by a vertical line (the "stem") and a marker at the end of the line indicating the actual data value.

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y)
plt.show()
```

```python
#linefmt parameter is used to change the style of stemps of plot

x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y, linefmt = "--")
plt.show()
```

```python
#marker paremter is used to change the style of marker 
#it can take two parameter:- first ic colour of marker and second style of marker

x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y, markerfmt = "r+")
plt.show()
```

```python
#bottom parameter is used to change the starting point of plot

x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y, bottom = 3)
plt.show()
```

```python
#basefmt parameter is used to change the color of base

x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y, basefmt = "r" )
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y, label = "Python")
plt.legend()
plt.show()
```

```python
x = [1,2,3,4,5,6]
y = [2,2,5,7,3,8]

plt.stem(x,y, orientation = "horizontal")
plt.show()
```
