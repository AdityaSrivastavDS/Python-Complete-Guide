```python
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\PC\\Downloads\\JamesBondMovies.csv")
```

```python
data
```

**dropping data**

```python
#using dropna() to drop all NaN values present within dataset
data.dropna()
```

```python
#dropping Nan along column
data.dropna(axis = 1)
```

```python
#removing an entire row which is NaN
data.dropna(how='any')
```

***subset***

```python
#removing NaN values from a specific column
#takes name oc column as parameter
data.dropna(subset=["Budget"])
```

```python
#remove NaN values and create a new dataset
data.dropna(inplace=True)

data
```

```python
#Will drop the row if it have only one NaN
#If it have multiple NaN values then it will not make any change

data.dropna(thresh = 1)
```

**fillna**

Used to fill NaN values



```python
#parameter is the value which we want to fill in place of NaN values
data.fillna(251)
```

```python
#filling values within specific column
#we will pass dictionary
#parameter:- column name as key and value which we want to fill
data.fillna({"Budget":47,"Box Office":278})
```

```python
#If we want to fill NaN values from its next value
data.ffill()
```

```python
#If we want to fill NaN values from its previos value
data.bfill()
```

```python
data.fillna(42,inplace = True)

data
```

```python
#using limit to fill data within restriction
data.fillna(67, limit=3)
```

**Replace**

```python
#replace() is used to replace existing value from dataset
#Parameter: to_replace() : value which we want to replace: and value = value with which we wanted to replace
data.replace(to_replace="Daniel Craig",value = "Daniel")
```

```python
'''regex is used to replace values alphabetical wise
Parameters: 1.|range of alphabet such as |A-B| or |c-f| take is whole as string
            2. The value with which we wanted to replace
            3. regex = True'''
data.replace("|A-Za-z|","Max",regex=True)
```

```python
#using dictionary to replace
data.replace({"Actor":'|A-C|'},'Sean Cody',regex = True)
```

**replacing values from backward and forward values**

```python
data.replace(456,method='ffill')
```

***limit***

```python
data.replace(to_replace='Sean Connery',value='Sean',limit = 5)
```

**Interpolate**

deals with blank data

In Pandas, the interpolate function is used to fill in missing or NaN (Not a Number) values in a DataFrame or Series. It provides a way to estimate values based on the existing data.


The interpolate function works by filling in the missing values with intermediate values that are computed based on the surrounding data points. It uses various interpolation methods to make these estimates. Some common interpolation methods include linear interpolation, polynomial interpolation, and spline interpolation.


```python
data.interpolate()
```

***methods of interpolate()***

```python
#along column
data.interpolate(method='linear',axis=0)
```

```python
#we can't interpolate data row wise as for that whole row must have same data type
```

```python
#using limit 
data.interpolate(limit = 10)
```

```python
#using limit along particular direction
data.interpolate(limit_direction='forward',limit=5)
```

```python
data.interpolate(limit_direction='backward',limit=5)
```

```python
#will limit from both forward and backward direction
data.interpolate(limit_direction='both',limit=5)
```
