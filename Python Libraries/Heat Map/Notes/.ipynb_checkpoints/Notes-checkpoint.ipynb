**Heatmap**

•A heatmap is a graphical representation of data where values in a matrix are represented as colors. 

•Seaborn, a statistical data visualization library in Python, provides a simple and effective way to create heatmaps. 

•Heatmaps are useful for visualizing relationships between two variables in a matrix, making it easy to identify patterns and trends in the data.

**Parameters and Function of heatmap**

1)data (array-like): 2D dataset that forms the matrix for the heatmap.


2)annot (bool, default: False): If True, write the data value in each cell.


3)cmap (str or Colormap, default: "viridis"): Colormap to define the colors of the heatmap.


4)linewidths (float, default: 0.5): Width of the lines that will divide each cell.


5)square (bool, default: False): If True, force the aspect ratio of the plot to be equal.


6)xticklabels, yticklabels ({"auto", bool, list-like}, default: "auto"): Control the display of axis labels.


7)cbar (bool or ax, default: True): Whether to draw a colorbar or not. You can also pass an Axes in which the colorbar will be drawn.


8)cbar_kws (dict, optional): Additional keyword arguments for the colorbar.


9)ax (matplotlib Axes, optional): Specifies the Axes on which the heatmap will be drawn.


10)fmt (str, optional): String formatting code to use when adding annotations.


11)linewidth (float, optional): Width of the lines that will divide each cell.


12)mask (boolean 2D array, optional): Mask to indicate "bad" values in the input data.
