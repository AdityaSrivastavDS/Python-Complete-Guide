```python
import matplotlib.pyplot as plt
```

```python
#plotting pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x)
plt.show()
```

```python
#labels parameter is used for labelling plot
#remember that labels is only used when we have multiple label

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

plt.pie(x, labels = y)

plt.show()
```

```python
#explode parameter is used to separate pies of pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

ex = [0.2,0.0,0.0,0.0]

plt.pie(x, labels = y, explode = ex)

plt.show()
```

```python
#cols parameter to change colour of pies

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col)

plt.show()
```

```python
#autopct = "%0.1f%%" is used to denote the percentage of each pie of pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, autopct = "%0.1f%%")

plt.show()
```

```python
#shadow = True parameter is used to create a shadow of the pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","orange","green"]

plt.pie(x, labels = y, colors = col, shadow = True)

plt.show()
```

```python
#radius parameter is used to change the size of pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","orange","green"]

plt.pie(x, labels = y, colors = col, radius = 0.8)

plt.show()
```

```python
#labeldistance parameter is used to change distance of label from the pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","orange","green"]

plt.pie(x, labels = y, colors = col, labeldistance = 1.3)

plt.show()
```

```python
#startangle parameter is used to change the starting point of pie chart
#We can also use it to rotate pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","orange","green"]

plt.pie(x, labels = y, colors = col, startangle = 90)

plt.show()
```

```python
#textprops is a parameter which is used to manipulate the text part of pie chart
#textprops have many attributes from which one is fontsize
#the attributes of textprops is always passed in dictionary format

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","orange","green"]

plt.pie(x, labels = y, colors = col, textprops = {"fontsize":15})

plt.show()
```

```python
#counterclock parameter is used to rotate the pie chart clockwise and anti-clockwise

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, counterclock = False)

plt.show()
```

```python
#wedgeprops is a parameter which is used to manipulate the edge part of pie chart
#wedgeprops have many attributes from which two are linewidth which is used to change the width of edge lines
#and width is used to increase the size of circle
#the attributes of wedgeprops is always passed in dictionary format


x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, shadow = True, wedgeprops = {"linewidth":8, "width":6})

plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, shadow = True, wedgeprops = {"width":6})

plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, wedgeprops ={"edgecolor":"black"})  #edgecolor attrubute changes the color of edge

plt.show()
```

```python
#center parameter is used to change the position of pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, center = (3,6))

plt.show()
```

```python
#rotatelabels parameter is used to rotate the labels of pie chart

x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col, rotatelabels = True)

plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col)

plt.title("Programming Language")   #for title of the plot

plt.show()
```

```python
x = [10,20,30,40]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, colors = col)

plt.title("Programming Language")

plt.legend(loc = 2)   #loc is used to change the location of legend

plt.show()
```

```python
#creating a dot pie chart

plt.pie([1])
plt.show()
```

```python
#plotting multiple pie charts in nested form

x = [10,20,30,40]
x1= [40,30,20,10]
y = ["C","C++","Java","Python"]

col = ["red","magenta","blue","yellow"]

plt.pie(x, labels = y, radius = 1.5)
plt.pie(x1, radius = 1, colors = col)

plt.show()
```

**Plotting Donut Chart:**

```python
#first way

x = [10,20,30,40]
x1= [40,30,20,10]
y = ["C","C++","Java","Python"]


plt.pie(x, labels = y, radius = 1.5)
plt.pie([1], colors = "white")

plt.show()
```

```python
#way 2

x = [10,20,30,40]
x1= [40,30,20,10]
y = ["C","C++","Java","Python"]


plt.pie(x, labels = y, radius = 1.5)

cr = plt.Circle(xy = (0,0), radius = 1, facecolor = "w")
plt.gca().add_artist(cr)


plt.show()
```
