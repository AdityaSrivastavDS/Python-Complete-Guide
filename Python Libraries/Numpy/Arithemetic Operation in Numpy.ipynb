```python
import numpy as np
```

**Arithmetic Operation**

***1-D Array:-***

***Addition***

```python
var = np.array([1,2,3,4])

varadd = var + 3 #3 will be added to every element of array

print(varadd)
```

```python
#addition of two arrays
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varadd = var1 + var2

print(varadd)
```

***Subtraction***

```python
var = np.array([1,2,3,4])

varsub = var - 3 #3 will be subtracted from every element of array

print(varsub)
```

```python
#subtraction of two arrays
var1 = np.array([1,2,3,4])
var2 = np.array([1,21,34,15])

varsub = var1 - var2

print(varsub)
```

***Multiplication***

```python
var = np.array([1,2,3,4])

varmul = var * 3 #3 will be muiltiplied from every element of array

print(varmul)
```

```python
#multiplication of two arrays
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

varmul = var1 * var2

print(varmul)
```

***Modulus***

```python
var = np.array([1,2,3,4])

vardiv = var % 3 #3 will be divided from every element of array

print(vardiv)
```

```python
#division of two arrays
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

vardiv = var2 % var1

print(vardiv)
```

**Using Arithmetic functions of numpy:-**

***add() function***

```python
#add() for addition
var1 = np.array([1,2,3,4])
var2 = np.array([5,6,7,8])

vardadd = np.add(var1,var2)

print(varadd)
```

***min and max***

uses to get maximum and minimum value in numpy array

```python
var = np.array([1,2,3,4,5,6])

print("Minimum value:- ", np.min(var))

print("Maximum value:- ", np.max(var))
```

***argmin and argmax***

these two functions are used to get index of minimum and maximum values of a numpy arrays

```python
var = np.array([1,2,3,4,5,6])

print("Minimum value:- ", np.min(var)," at index " ,np.argmin(var))

print("Maximum value:- ", np.max(var)," at index " ,np.argmax(var))
```

**Arithmetic functions for 2-D array:-**

Numpy array has two axis where axis = 0 denotes column while axis = 1 represents row

```python
#seeking minimum value in row
var1 = np.array([[1,2,3],[4,5,6]])

print(np.min(var1,axis=1))
```

```python
#seeking minimum value in column
var1 = np.array([[1,2,3],[4,5,6]])

print(np.min(var1,axis=0))
```

***sqrt***

Used to calculate squrae root

```python
var = np.array([8,3,6,7,5,6])

print(np.sqrt(var))
```

```python
#sin function
var2 = np.array([1,2,3])

print(np.sin(var2))
```

```python
#cos function
var2 = np.array([1,2,3])

print(np.cos(var2))
```

```python
#cumsum function
var2 = np.array([1,2,3])

print(np.cumsum(var2))
```
