**Box Plot:**

â€¢A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It provides a summary of key statistical measures and helps identify the central tendency, spread, and potential outliers within the data. 


***1)Central Tendency:*** The box in the plot represents the interquartile range (IQR), with the line inside the box indicating the median (middle value) of the dataset. This gives a sense of the central tendency of the data.


***2)Spread of the Data:*** The length of the whiskers (lines extending from the box) provides information about the spread or variability of the data. Whiskers typically extend up to a certain multiple of the IQR.


***3)Outliers:*** Individual data points that fall outside the whiskers are considered outliers and are often plotted individually. Box plots help identify and visualize these potential outliers.


***4)Skewness:*** The asymmetry of the data distribution can be visually assessed by observing the relative lengths of the whiskers and the position of the median.


***5)Comparisons:*** Box plots are useful for comparing the distribution of multiple datasets or groups. Multiple boxes can be plotted side by side for easy comparison.


***6)Distributions:*** Box plots provide a visual representation of the overall shape of the data distribution, giving an idea of whether the data is symmetric, skewed, or has multiple modes.

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x)
plt.show()
```

```python
#notch = True parameter is used to change the style of box

x = [10,20,30,40,50,60,70]

plt.boxplot(x, notch = True)
plt.show()
```

```python
#vert = False will plot the box plot in horizontal 

x = [10,20,30,40,50,60,70]

plt.boxplot(x, vert = False)
plt.show()
```

```python
#widths parameter is used to change the width of box plot

x = [10,20,30,40,50,60,70]

plt.boxplot(x, widths = 0.8)
plt.show()
```

```python
x = [10,20,30,40,50,60,70]

plt.boxplot(x, labels = ["Python"])
plt.show()
```

```python
#patch_artist =True will fill the color inside the box plot

x = [10,20,30,40,50,60,70]

plt.boxplot(x, patch_artist = True)
plt.show()
```

```python
#showmeans = True will show the mean of the dataset

x = [10,20,30,40,50,60,70]

plt.boxplot(x, showmeans = True)
plt.show()
```

```python
#representation of outlier

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x)
plt.show()
```

```python
#whis parameter is used to join the stem of box plot at top

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, whis = 2)
plt.show()
```

```python
#sym parameter is used to change the color and style of outlier

x = [10,20,30,40,50,60,70,120]

plt.boxplot(x, sym = "r+")
plt.show()
```

```python
#plotting multiple box plots within same plot

x = [10,20,30,40,50,60,70]
x1 = [10,30,20,50,80,70,40]

#y is a list that stores two list 
y = [x,x1]

plt.boxplot(y, labels=["Python","Java"])
plt.show()
```
