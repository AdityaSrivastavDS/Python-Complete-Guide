**Data Type**

```python
import numpy as np
```

```python
var = np.array([1,2,3,4])

print("Data Type:- ",var.dtype)
```

```python
var = np.array([1,2,3,4,12,16,18])

print("Data Type:- ",var.dtype)
```

```python
var = np.array([1.2,1.4,1.6])

print("Data Type:- ",var.dtype)
```

```python
var = np.array(['a','b','c'])

print("Data Type:- ",var.dtype)
```

```python
var = np.array(['a','b','c',1,2,3])

print("Data Type:- ",var.dtype)
```

***Converting Data Type:-***

```python
x = np.array([1,2,3,4])

print("Data Type:- ",x.dtype)
```

```python
#using data type for conversion
x = np.array([1,2,3,4],dtype = np.int8)

print(x)
print("Data Type:- ",x.dtype)
```

```python
#using short keys for data conversion
x1 = np.array([1,2,3,4],dtype = "f")

print(x1)
print("Data Type:- ",x1.dtype)
```

***Conversion of Data Type as a function:-***

```python
x2 = np.array([1,2,3,4])
new = np.float32(x2)


print("Data Type:- ",x2.dtype)
print("Data Type:- ",new.dtype)
print(x2)
print(new)
```

```python
#using astype() funtion for data conversion
x3 = np.array([1,2,3,4])

new1 = x3.astype(float)

print(x3)
print(new1)
```
