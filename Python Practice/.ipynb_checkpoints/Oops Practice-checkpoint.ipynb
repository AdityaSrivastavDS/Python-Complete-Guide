```python
#simple calculator
class calculator:
    def __init__(self,a,b):
        self.a = a
        self.b = b
    def addition(self):
        return self.a+self.b
    def subtraction(self):
        return self.a-self.b
    def division(self):
        return self.a/self.b
    def product(self):
        return self.a*self.b
    
a = int(input("Enter number 1: "))
b = int(input("Enter number 2: "))

#taking operator input
opr = input('''
            1. + for addition
            2. - for subtraction
            3. / for devision
            4. * for multiplication''')

#creating object of class calculator
obj = calculator(a,b)


#defining conditions using match
match(opr):
    case '+':
        print(obj.addition())
    case '-':
        print(obj.subtraction())
    case '/':
        print(obj.division())
    case '*':
        print(obj.product())
```

```python
#If we dont wanted to write self again and again in our methods then we can simply declare them as static methods
class Calculator:
    def __init__(self, a, b):
        self.a = a
        self.b = b

    @staticmethod
    def addition(a, b):
        return a + b

    @staticmethod
    def subtraction(a, b):
        return a - b

    @staticmethod
    def division(a, b):
        return a / b

    @staticmethod
    def product(a, b):
        return a * b

a = int(input("Enter number 1: "))
b = int(input("Enter number 2: "))

# Taking operator input
opr = input('''
1. + for addition
2. - for subtraction
3. / for division
4. * for multiplication
''')


# Defining conditions using match-case
match opr:
    case '+':
        print(Calculator.addition(a, b))
    case '-':
        print(Calculator.subtraction(a, b))
    case '/':
        print(Calculator.division(a, b))
    case '*':
        print(Calculator.product(a, b))
    case _:
        print("Invalid operation")
```

# Class and Instance(object) Attribute

We have two instances:

***1)class attribute***   
***2)object attribute***


Class attribute is common to entire class while object attribute is object specific and does not apply to every object of the class.
    

**Take an Example:**

```python
class student:
    def __init__(self,name):
        self.name = name
obj = student("Ishika")

print(obj.name)
```

Above we have a class named as student in which name is a class attribte which is common as every student will have some name

```python
'''Ques. Create a student class that takes names and marks of three subjects as arguments in constructor.
Then create a method to print average'''

class student:
    def __init__(self,name,marks):
        self.name = name
        self.marks = marks
    def calc_avg(self):
        sums = 0
        for mark in self.marks:
            sums+=mark
        print("Hii,",self.name," your average score is ",sums/3)

subj1 = student("Rohit",[83,45,67])
subj2 = student("Manasvi",[86,67,87])
subj3 = student("Alia",[68,45,67])


print(subj1.calc_avg())
print(subj2.calc_avg())
print(subj3.calc_avg())
```

# Method Overloading and Overriding 

### Method overloading

Method overloading is the ability to define multiple methods with the same name but different parameters. Unlike some other programming languages, Python does not support method overloading in the traditional sense. Instead, you can achieve similar functionality using default arguments or variable-length arguments.

```python
#Using Default Arguments
class MathOperations:
    def add(self, a, b=0, c=0):
        return a + b + c

math_ops = MathOperations()
print(math_ops.add(5))        # Outputs: 5
print(math_ops.add(5, 10))    # Outputs: 15
print(math_ops.add(5, 10, 15)) # Outputs: 30
```

### Method Overriding

Method overriding allows a subclass to provide a specific implementation of a method that is already defined in its superclass. This is useful when you want to change or extend the behavior of a method in the parent class.

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):  # Overriding the speak method
        print("Dog barks")

class Cat(Animal):
    def speak(self):  # Overriding the speak method
        print("Cat meows")

# Creating objects of the child classes
dog = Dog()
cat = Cat()

dog.speak()  # Outputs: Dog barks
cat.speak()  # Outputs: Cat meows
```

```python
#using super method to call method of super or parent class
class Animal:
    def speak(self):
        print("The animal produces sond")
        
class Pet(Animal):
    def dog(self):
        #calling speak() method of Animal class using super()
        super().speak()
        print("The dog barks!")
        
obj = Pet()

obj.dog()
```

# Inheritance

Inheritance is a mechanism that allows one class (the child class) to inherit attributes and methods from another class (the parent class). There are several types of inheritance in Python:

### 1.Single Inheritance

Single inheritance involves a single parent class and a single child class.

```python
#parent class
class animal:
    def speak():
        print("A animal produce sounds")
        
#child class 
class dog(animal):
    def bark():
        print("The dog barks!")
        

obj = dog()

dog.speak()
dog.bark()
```

### 2.Multiple Inheritance

Multiple inheritance involves multiple parent classes and a single child class.

```python
#parent class 1
class Animal:
    def types(self):
        print("Animal have different categories")

#parent class 2
class Pet:
    def domestic(self):
        print("Dog is a domestic animal")
        
#child class
class Dog(Animal, Pet):
    def quality(self):
        print("The dogs are calm and really pure hearted animals")
        
obj = Dog()

obj.types()
obj.domestic()
obj.quality()
```

### 3.Multilevel Inheritance

Multilevel inheritance involves a chain of classes where a class is derived from another derived class.

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal speaks")

# Intermediate child class
class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

# Child class
class Dog(Mammal):
    def bark(self):
        print("Dog barks")

# Creating an object of the child class
dog = Dog()
dog.speak()  # Inherited from Animal
dog.walk()   # Inherited from Mammal
dog.bark()   # Child class method
```

### 4.Hierarchical Inheritance

Hierarchical inheritance involves multiple child classes inheriting from a single parent class.

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal speaks")

# Child class 1
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Child class 2
class Cat(Animal):
    def meow(self):
        print("Cat meows")

# Creating objects of the child classes
dog = Dog()
cat = Cat()

dog.speak()  # Inherited method
dog.bark()   # Child class method

cat.speak()  # Inherited method
cat.meow()   # Child class method
```

###  5.Hybrid Inheritance

Hybrid inheritance is a combination of two or more types of inheritance. It can be a mix of single, multiple, hierarchical, or multilevel inheritance

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal speaks")

# Intermediate class 1
class Mammal(Animal):
    def walk(self):
        print("Mammal walks")

# Intermediate class 2
class Pet:
    def show_affection(self):
        print("Pet shows affection")

# Child class
class Dog(Mammal, Pet):
    def bark(self):
        print("Dog barks")

# Creating an object of the child class
dog = Dog()
dog.speak()          # Inherited from Animal
dog.walk()           # Inherited from Mammal
dog.show_affection() # Inherited from Pet
dog.bark()           # Child class method
```

## Intermediate to Advance Unique concepts of Python's Inheritance 

### Diamond Problem in Python

The diamond problem occurs when a class inherits from two classes that have a common base class. Python handles this using the MRO.

```python
class A:
    def method(self):
        print("Method of class A")

class B(A):
    def method(self):
        print("Method of class B")
        super().method()
        
class C(A):
    def method(self):
        print("Method of class C")
        super().method()
    
class D(B, C):
    def method(self):
        print("Method of Class D")
        super().method()
        
obj = D()
obj.method()

print()

print(D.mro())
```

**Note:-**

In last line of above code, MRO stands for Method Resolution Order. It is the order in which Python looks for a method in a hierarchy of classes. When a method is called on an instance of a class, Python uses the MRO to determine the order in which to search for that method in the class and its base classes.

# Abstraction

Abstraction is one of the fundamental concepts in object-oriented programming (OOP). It involves hiding the complex implementation details and showing only the necessary features of an object. Abstraction is typically implemented using abstract classes and interfaces.

In Python, abstraction can be achieved using the abc module, which provides the infrastructure for defining abstract base classes (ABCs).

<img src="https://www.codingem.com/wp-content/uploads/2021/10/pyimg.001.jpeg" alt="Description" width="600" height="500">

```python
from abc import ABC, abstractmethod

class Animal:
    @abstractmethod
    def sound(self):
        pass
    
class Dog(Animal):
    def sound(self):
        return "bark"

class Cat(Animal):
    def sound(self):
        return "meow"
    
dog = Dog()
cat = Cat()

print(dog.sound())
print(cat.sound())
```

```python
#Intermediate Implementation

from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
    
    @abstractmethod
    def perimeter(self):
        pass
    
class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def area(self):
        return self.width * self.height
    
    def perimeter(self):
        return 2*(self.width + self.height)
    
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
        
    def area(self):
        return 3.14 *self.radius *2
    
    def perimeter(self):
        return 2 * 3.14 *self.radius
    

#creating instances of the subclasses
rect = Rectangle(5, 10)
circ = Circle(9)
    

print("Rectangle Area: ",rect.area())
print("Perimeter of rectangle: ",rect.perimeter())

print()

print("Area of circle: ",circ.area())
print("Perimeter of circle: ",circ.perimeter())
```

```python
#Advance implementation

from abc import ABC, abstractmethod

class Vehicle(ABC):
    def __init__(self, make, model):
        self.make = make
        self.model = model

    @abstractmethod
    def start_engine(self):
        pass

    @abstractmethod
    def stop_engine(self):
        pass

    def get_info(self):
        return f"Vehicle: {self.make} {self.model}"

class Car(Vehicle):
    def start_engine(self):
        return f"{self.make} {self.model}'s engine started"

    def stop_engine(self):
        return f"{self.make} {self.model}'s engine stopped"

class Motorcycle(Vehicle):
    def start_engine(self):
        return f"{self.make} {self.model}'s engine started"

    def stop_engine(self):
        return f"{self.make} {self.model}'s engine stopped"

# Creating instances of the subclasses
car = Car("Toyota", "Corolla")
motorcycle = Motorcycle("Honda", "CBR600")

print(car.get_info())  # Outputs: Vehicle: Toyota Corolla
print(car.start_engine())  # Outputs: Toyota Corolla's engine started
print(car.stop_engine())   # Outputs: Toyota Corolla's engine stopped

print()

print(motorcycle.get_info())  # Outputs: Vehicle: Honda CBR600
print(motorcycle.start_engine())  # Outputs: Honda CBR600's engine started
print(motorcycle.stop_engine())   # Outputs: Honda CBR600's engine stopped
```

# Polymorphism

Polymorphism is one of the core concepts in object-oriented programming. It allows objects of different classes to be treated as objects of a common superclass. The most common use of polymorphism is when a parent class reference is used to refer to a child class object. Polymorphism simplifies code and enhances flexibility and maintainability by allowing you to write generic code that can work with objects of various types.

<img src="https://th.bing.com/th/id/OIP.EmW7lJFjUeq6aYCGHIin6AHaCp?rs=1&pid=ImgDetMain" alt="Description" width="600" height="500">

```python
#basic implementation
#Polymorphism with Method Overriding

class Animal:
    def sound(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

def make_sound(animal: Animal):
    print(animal.sound())

# Creating instances of subclasses
dog = Dog()
cat = Cat()

# Demonstrating polymorphism
make_sound(dog)  # Outputs: Bark
make_sound(cat)  # Outputs: Meow
```

***In above example, make_sound is a function that accepts an object of type Animal. Both Dog and Cat classes override the sound
method, demonstrating polymorphism as make_sound can handle different types of Animal objects.***

### Intermediate Polymorphism

**Example: Polymorphism with Duck Typing**

Duck typing is a concept related to polymorphism where the type or class of an object is less important than the methods it defines. If an object implements the necessary methods, it can be used regardless of its class.

```python
class Dog:
    def sound(self):
        return "Bark"

class Cat:
    def sound(self):
        return "Meow"

class Bird:
    def sound(self):
        return "Chirp"

def make_sound(animal):
    print(animal.sound())

# Creating instances
dog = Dog()
cat = Cat()
bird = Bird()

# Demonstrating polymorphism with duck typing
make_sound(dog)  # Outputs: Bark
make_sound(cat)  # Outputs: Meow
make_sound(bird)  # Outputs: Chirp
```

### Advanced Polymorphism

**Example: Polymorphism with Abstract Base Classes (ABC)**

Using abstract base classes to enforce polymorphism ensures that all subclasses implement specific methods.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass

class Rectangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def area(self):
        return self.width * self.height

    def perimeter(self):
        return 2 * (self.width + self.height)

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return 3.14 * self.radius ** 2

    def perimeter(self):
        return 2 * 3.14 * self.radius

def print_shape_info(shape: Shape):
    print(f"Area: {shape.area()}")
    print(f"Perimeter: {shape.perimeter()}")

# Creating instances of subclasses
rectangle = Rectangle(5, 10)
circle = Circle(7)

# Demonstrating polymorphism with abstract base classes
print_shape_info(rectangle)

print_shape_info(circle)
```

***In above example, Shape is an abstract base class that defines the interface for its subclasses. Rectangle and Circle implement the required methods. The print_shape_info function works with any Shape object, demonstrating polymorphism.***

### Example: Real-world Use Case - Drawing Shapes

Consider a drawing application where you can draw various shapes. Using polymorphism, you can define a common interface for all shapes and write a generic drawing function.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def draw(self):
        pass

class Circle(Shape):
    def draw(self):
        print("Drawing a circle")

class Square(Shape):
    def draw(self):
        print("Drawing a square")

class Triangle(Shape):
    def draw(self):
        print("Drawing a triangle")

def draw_shape(shape: Shape):
    shape.draw()

# Creating instances of different shapes
shapes = [Circle(), Square(), Triangle()]

# Drawing all shapes
for shape in shapes:
    draw_shape(shape)
```
