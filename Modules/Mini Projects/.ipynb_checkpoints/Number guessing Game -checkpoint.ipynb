```python
import random
#initializing game starter
player1 = random.randint(1,101)

#taking user input
player2 = int(input("Enter a random number:- "))

#displaying the random value generated by the system
print('The player1 random number is:- ',player1)

#setting conditions and rules for the game
if player1 > player2:
    print("Poor of you! Try Again")
    
elif player1<player2:
    print("Your Guess is Too High")
    
else:
    print("Congrats!You Won")
    
```
