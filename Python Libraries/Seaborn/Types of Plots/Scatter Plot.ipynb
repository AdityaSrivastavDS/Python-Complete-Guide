```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("E:\\Aditya\\Datasets\\Automobile_data.csv")
```

```python
data.head()
```

```python
#plotting scatterplot

plt.figure(figsize = (10,10))

sns.scatterplot(x = "engine-size", y = "wheel-base", data = data)

plt.show()
```

```python

#changing color of data points using color parameter
plt.figure(figsize = (10,10))

sns.scatterplot(x = "engine-size", y = "wheel-base", color = "red", data = data)

plt.show()
```

```python
#changing shape of data points using marker parameter
plt.figure(figsize = (10,10))

sns.scatterplot(x = "engine-size", y = "wheel-base", color = "red", marker = ">", data = data)

plt.show()
```

```python
#show groups with different colors using hue
plt.figure(figsize = (10,10))

sns.scatterplot(x = "engine-size", y = "wheel-base", color = "red",hue = "fuel-type", data = data)

plt.show()
```

```python
#setting limits for x axis
plt.figure(figsize = (10,10))

plt.xlim([50,200])

sns.scatterplot(x = "engine-size", y = "wheel-base", color = "red", data = data)

plt.show()
```

```python
#showing two different grouping variables using hue and style
plt.figure(figsize = (10,10))

sns.scatterplot(x = "engine-size", y = "wheel-base",  hue = "fuel-type", style = "num-of-cylinders", data = data)

plt.show()
```

```python
#changing sizes
plt.figure(figsize = (10,10))

sns.scatterplot(x = "engine-size", y = "wheel-base", hue = "fuel-type", style = "num-of-cylinders",size = "city-mpg",sizes = (20,200), data = data)

plt.show()
```
