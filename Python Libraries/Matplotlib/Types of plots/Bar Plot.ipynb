```python
import matplotlib.pyplot as plt
```

```python
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.bar(x,y)

plt.show()
```

```python
'''1.xlabel is used for labelling x axis
   2.ylabel is used for labelling y axis
   3.title is used for giving heading or title to the graph'''


x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y)

plt.show()
```

```python
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

#fontsize is used to change the font size of plot
plt.xlabel("language", fontsize = 15)
plt.ylabel("No", fontsize = 15)
plt.title("Python")

plt.bar(x,y)

plt.show()
```

```python
#changing width of bars using width parameter
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, width = 0.5)

plt.show()
```

```python
#changing colour of bars
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, color = "red")

plt.show()
```

```python
#giving multiple colours to each bars
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

#defining colours for each bar
col = ["blue", "purple", "magenta", "purple"]

plt.bar(x,y, color = col)

plt.show()
```

```python
#CHANGING ALIGNMENT OF BARS USING ALIGN PARAMETER
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, align = "edge")

plt.show()
```

```python
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, align = "center")

plt.show()
```

```python
#CHANGING COLOUR OF EDGE OF BARS USING EDGE PARAMETER
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, edgecolor = "red")

plt.show()
```

```python
#CHANGING WIDTH OF EDGE
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, edgecolor = "red", linewidth = 5)

plt.show()
```

```python
#CHANGING LINESTYTLE OF EDGE USING linestyle parameter

x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, edgecolor = "red", linewidth = 5, linestyle=":")

plt.show()
```

```python
#CHANGING THE BRIGHTNESS OF THE COLOUR OF THGE BAR using alpha parameter
x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, alpha = 0.5)

plt.show()
```

```python
#CREATING LABEL OF THE GRAPH 

x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, label = "Programming Languages")

#must when passing label
plt.legend()

plt.show()
```

```python
#PLOTTING MULTPLE OVERLAPPED BAR GRAPHS WITHIN SAME PLOT

x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

plt.bar(x,y, color = "red", label = "Popularity")
plt.bar(x,z, color = "yellow", label = "Popularity1")


plt.legend()

plt.show()
```

```python
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

#creating array equal to length of x
p = np.arange(len(x))

p1 = [j+width for j in p]



plt.bar(p,y, color = "red", label = "Popularity")
plt.bar(p1,z, color = "yellow", label = "Popularity1")

plt.xticks(p+width%2, x)

plt.legend()

plt.show()
```

```python
x = ["Python","C","C++","Java"]
y = [85,70,60,82]
z = [20,30,40,50]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

#creating array equal to length of x
p = np.arange(len(x))

p1 = [j+width for j in p]



plt.bar(p,y, color = "red", label = "Popularity")
plt.bar(p1,z, color = "yellow", label = "Popularity1")

#rotating the labels using rotation parameter
plt.xticks(p+width%2, x, rotation = 10)

plt.legend()

plt.show()
```

```python
#CREATINTG  HORIZONTAL GRAPH USING barh

x = ["Python","C","C++","Java"]
y = [85,70,60,82]

plt.xlabel("language")
plt.ylabel("No")
plt.title("Python")

#defining colours for each bar
col = ["blue", "purple", "magenta", "red"]

plt.barh(x,y, color = col, edgecolor = "black")

plt.show()
```
