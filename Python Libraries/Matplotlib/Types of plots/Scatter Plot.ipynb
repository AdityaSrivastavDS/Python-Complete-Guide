```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,5,6,1,9]

plt.scatter(x,y)
plt.show()
```

```python
#title is used to give title to graph
#fontsize is used to increase size of font

x = [1,2,3,4,5,6,7]
y = [2,3,4,5,6,1,9]

plt.scatter(x,y)
plt.title("Scatter Plot", fontsize = 15)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,5,6,1,9]

plt.scatter(x,y)
plt.xlabel("Day")
plt.ylabel("No")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [2,3,4,5,6,1,9]

col = [10,40,28,53,98,12,75]
sizes = [400,600,200,300,500,700,400]

plt.figure(figsize = (20,10))

plt.scatter(x,y, c = col , s = sizes, cmap = "Pastel1")

t = plt.colorbar()

t.set_label("Colour Bar", fontsize = 10)

plt.xlabel("Day")
plt.ylabel("No")

plt.show()
```
