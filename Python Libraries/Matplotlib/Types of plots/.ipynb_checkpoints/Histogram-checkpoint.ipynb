```python
num = [15,43,10, 17 ,27 ,48 ,39 ,14 ,49 ,46 ,16 ,44 ,32 ,59 ,36 ,22, 19 ,48 ,40 ,40 ,40 ,29 ,11 ,32,
 50, 41 ,24 ,45, 35, 57 ,33, 24 ,30 ,52, 40 ,40 ,44 ,20 ,18 ,56, 31 ,32 ,38, 33 ,59, 26 ,36, 54,
 14, 43]

plt.hist(num, color = "magenta", edgecolor = "black", label = "Programming language")

plt.title("Numeric Data")

plt.xlabel("Python")
plt.ylabel("Popularity")

plt.legend()

plt.show()
```

```python
#axvline() is used to divide the histogram 
#It takes a value range present within graph or in simple terms the point at which we want to divide the graph

num = [15,43,10, 17 ,27 ,48 ,39 ,14 ,49 ,46 ,16 ,44 ,32 ,59 ,36 ,22, 19 ,48 ,40 ,40 ,40 ,29 ,11 ,32,
 50, 41 ,24 ,45, 35, 57 ,33, 24 ,30 ,52, 40 ,40 ,44 ,20 ,18 ,56, 31 ,32 ,38, 33 ,59, 26 ,36, 54,
 14, 43]

plt.hist(num, color = "magenta", edgecolor = "black", label = "Programming language")

plt.title("Numeric Data")

plt.xlabel("Python")
plt.ylabel("Popularity")

plt.axvline(50, color = "blue", label = "axvline")

plt.legend()

plt.show()
```

```python
#grid(True) is used to deploy grid to graph

num = [15,43,10, 17 ,27 ,48 ,39 ,14 ,49 ,46 ,16 ,44 ,32 ,59 ,36 ,22, 19 ,48 ,40 ,40 ,40 ,29 ,11 ,32,
 50, 41 ,24 ,45, 35, 57 ,33, 24 ,30 ,52, 40 ,40 ,44 ,20 ,18 ,56, 31 ,32 ,38, 33 ,59, 26 ,36, 54,
 14, 43]

plt.hist(num, color = "magenta", edgecolor = "black", label = "Programming language")

plt.title("Numeric Data")

plt.xlabel("Python")
plt.ylabel("Popularity")

plt.axvline(50, color = "blue", label = "axvline")

plt.legend()
plt.grid(True)

plt.show()
```
